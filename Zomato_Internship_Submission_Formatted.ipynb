# Zomato Restaurant Reviews - Sentiment Analysis Project

**Objective:** Build a machine learning model to classify customer reviews as Positive, Neutral, or Negative based on their text.

This notebook follows all steps mentioned in the submission template: Data Loading â†’ EDA â†’ Data Cleaning â†’ Visualization â†’ Feature Engineering â†’ Modeling â†’ Evaluation â†’ Conclusion.

## 1. Import Libraries

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

## 2. Load Dataset

```python

reviews_df = pd.read_csv("Zomato Restaurant reviews.csv")
meta_df = pd.read_csv("Zomato Restaurant names and Metadata.csv")

print("Reviews Shape:", reviews_df.shape)
print("Metadata Shape:", meta_df.shape)

reviews_df.head()
```

## 3. Data Understanding & EDA

```python

reviews_df.info()
reviews_df['Rating'].value_counts()
```

## 4. Data Cleaning

```python

# Remove missing values
reviews_df = reviews_df.dropna(subset=['Review', 'Rating'])

# Convert ratings to numeric
reviews_df['Rating'] = pd.to_numeric(reviews_df['Rating'], errors='coerce')
reviews_df = reviews_df.dropna(subset=['Rating'])

# Text cleaning function
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)
    return text

reviews_df['Cleaned_Review'] = reviews_df['Review'].apply(clean_text)

reviews_df.head()
```

## 5. Label Creation (Sentiment Classes)

```python

# Create sentiment labels based on rating
def get_sentiment(rating):
    if rating >= 4:
        return "Positive"
    elif rating == 3:
        return "Neutral"
    else:
        return "Negative"

reviews_df['Sentiment'] = reviews_df['Rating'].apply(get_sentiment)
reviews_df['Sentiment'].value_counts()
```

## 6. Data Visualization

```python

plt.figure(figsize=(6,4))
sns.countplot(x='Sentiment', data=reviews_df)
plt.title("Sentiment Distribution")
plt.show()
```

## 7. Feature Engineering

```python

X = reviews_df['Cleaned_Review']
y = reviews_df['Sentiment']

vectorizer = TfidfVectorizer(max_features=5000)
X_vec = vectorizer.fit_transform(X)
```

## 8. Train-Test Split

```python

X_train, X_test, y_train, y_test = train_test_split(
    X_vec, y, test_size=0.2, random_state=42, stratify=y
)
```

## 9. Model Training

```python

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

## 10. Model Evaluation

```python

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```
